```python
import tushare as ts
import pandas as pd
```

```python
pro = ts.pro_api()
pro = ts.pro_api('f7ee2dda50509f466b74a0eb3cbbe9a60b6c6cf16e3005949cf70672')
df = pro.daily(ts_code='600519.SH')
```

```python
df = df.sort_values('trade_date')
df['trade_date'] = pd.to_datetime(df['trade_date'])
df
```

```python
df.set_index('trade_date',inplace=True)
data = df
```

```python
data
```

```python
kl_data = data.values[:,[1,4,3,2]]
```

```python
data.index
```

```python
from pyecharts.charts import Kline
from pyecharts import options as opts
k_obj = Kline()
k_obj.add_xaxis(data.index.strftime("%Y-%m-%d").tolist())
k_obj.add_yaxis("贵州茅台-日K线图",kl_data.tolist())
k_obj.set_global_opts(yaxis_opts=opts.AxisOpts(is_scale=True), xaxis_opts=opts.AxisOpts(is_scale=True),title_opts=opts.TitleOpts(title=""))
k_obj.render("贵州茅台股票日K线图.html")
```
